```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line Plot


```python
enroll=pd.read_csv('Enrollment.csv')
enroll
```

```python
plt.figure(figsize=(8,5))
plt.xlabel("Serial Number")
plt.ylabel("Number of Students")
plt.title('State wise enrollment of class 1 boys and class 1 girls')
plt.plot(enroll.Serial_number,enroll.Enrolment_in_ClassI_Boys,label='class 1 Boys')
plt.plot(enroll.Serial_number,enroll.Enrolment_in_ClassI_Girls,label='class 1 Girls')
# plt.plot(enroll.STATE,enroll.)
plt.legend()
plt.show()
```

# Box Plot

```python
plt.boxplot([enroll.Enrolment_in_ClassI_Boys,enroll.Enrolment_in_ClassI_Girls])
plt.show()
```

# Scatter Plot

```python

plt.scatter(enroll.Enrolment_in_ClassI_Boys,enroll['Enrolment in Class I - Total'], s=10, c='green',edgecolor='black')
plt.xlabel("total no.of students enrolled in class 1")
plt.ylabel("No.of Boys enrolled in class 1")
plt.show()
```
